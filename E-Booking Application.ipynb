## Number of bookings per city

```python
values = [len(City_A), len(City_B), len(City_C), len(City_D), len(City_E)]
labels = ['City A', 'City B', 'City C', 'City D', 'City E']

def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n{:d}".format(pct, absolute)

colors = sns.color_palette('Greens_d')
plt.figure(figsize = (8,8))
plt.pie(values, labels = labels, colors = colors, autopct = lambda pct: func(pct, values), textprops={'fontsize': 16},wedgeprops={'linewidth': 3.0, 'edgecolor': 'white'})
plt.title('Number of bookings per city', fontsize=15)
plt.show()
```
